why `readline()` exists when we can just `read()`?

 - use readline when the file has a large amount of data and you don't want to load the whole thing in the memory at once, readline will load each line simulataneously saving the possible load on the memory.

```python
#  custom code to read the entire data using `readline()`
f=open('sample.txt','r')

while True:
  data=f.readline()

  if data=='':
    break

  else:
    print(data,end='')

f.close()
```

# **Context manager - `with`**

- when used while opening a file, the file is closed as soon as the work is done, thus freeing up resources (memory) and not comprimising in the safety of the files.

- tho if files not closed, they'll be closed eventually by the garbage collector

- an alternate to f.close()

```python
# with

with open('sample1.txt', 'w') as f:
  f.write('robert pattinson firse')
```

```python
# with closed the file, proof of it
f.write('henlo')
```

```python
with open('sample.txt','r') as f:
  print(f.read())
```

```python
# moving within a file 'n' character by 'n' character
with open('sample.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
```

```python
# benefit of moving within a file like mentioned above
# (helps load big memory consuming files)

big=['dilo ka shooter hai mera scooter\n' for i in range(1000)]

with open ('bigfile.txt','w') as f:
  f.writelines(big)
```

```python
with open('bigfile.txt','r') as f:
  chunk_size=10

  data=f.read(chunk_size)
  while len(data) > 0:
    print(data, end='--\n')
    data=f.read(chunk_size)
  print(f.tell())
```

# Serialization and Deserialization

- **serialization**: process of converting python data types to JSON format

- **deserialization**: process of converting JSON file to python files

**JSON - Java script object notation** UNIVERSAL format that every language understands

```python
# serialization using json module

# to write in this format - serialization
import json

d={'sanhita': 'saxena',
   'blah bla blah':'ek do teen chaar'}

with open('json_demo.json','w') as f:
  json.dump(d,f, indent=4)
```

```python
#  to read in this format - deserialization

with open('json_demo.json','r') as f:
  data=json.load(f)
  print(data)
  print(type(data))
```

```python
# serialize and deserialize a tuple

import json
t = (1,2,3,4,5,6)

with open('json_demo.json','w') as f:
  json.dump(t,f)

with open('json_demo.json','r') as f:
  data=json.load(f)
  print(data)
  print(type(data))
  print(type(f))
```

# serializing  and deserializing custom objects

```python
class person:

  def __init__(self, name, age, gender):
    self.name=name
    self.age=age
    self.gender=gender
```

```python
p=person('sanhita saxena', 25,'female')
```

```python
# to visualize this custom object

with open('json_demo.json','w') as f:
  json.dump(p,f)
  # will show error because json doesn't know how to print or store our file,
  # so we will have to create a separate method to specify how to store our file
```

storing custom method as a string

```python
# creating method that specifies how to store our custom objec

def store_object_as(p):
  if isinstance(p,person): #to check if the passed parameter is actually an instance of this class or no
    return "{} is {} years of age and is a {}.".format(p.name,p.age,p.gender)

# now we can write it in json using 'default' parameter
with open('json_demo.json','w') as f:
  json.dump(p,f, default=store_object_as)
```

storing custom method as a dictionary

```python
def store_object_as(p):
  if isinstance(p,person): #to check if the passed parameter is actually an instance of this class or no
    return {'name': p.name,
            'age': p.age,
            'gender': p.gender}

# now we can write it in json using 'default' parameter
with open('json_demo.json','w') as f:
  json.dump(p,f, default=store_object_as,indent=4)
```

```python
# deserialing

with open('json_demo.json','r') as f:
  d=json.load(f)
  print(d)
  print(type(d))
```

**You can't as it is store your custom object, in json file, that is you can't serialize and deserialize your custom objects**

To overcome the challenge mentioned above we convert our custom object to a binary format called pickling.

- `pickling` is a process where a python object hierarchy is converted to a byte stream

- `unpickling` is the inverse operation, where byte stream is converted back into object hierarchy.


```python
class person:

  def __init__(self, name, age, gender):
    self.name=name
    self.age=age
    self.gender=gender

  def display(self):
    print(f'hi my name is {self.name}, i am {self.age} yo {self.gender}.')

```

```python
p=person('sanhita',25,'girl')
```

```python
# pickle dump - serializing
import pickle

with open('person.pkl', 'wb') as f:
  pickle.dump(p,f)
```

```python
# pickle load - deserializing
import pickle

with open('person.pkl','rb') as f:
  data= pickle.load(f)
  print(data)
# p.display()
```

# difference between JSON and PICKLE

- we use JSON when we want to store the data in a human readable format

- we use pickle when we want to store our data in a binary format and preserve its functionality
